```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, MaxAbsScaler
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_learning_curves
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

x_test = pd.read_csv("../data/COMP30027_2021_Project2_datasets/recipe_test.csv", index_col = False, delimiter = ',', header=0)


x_test
```

```python
x_test['steps'] = x_test['steps'].apply(eval)
x_test['ingredients'] = x_test['ingredients'].apply(eval)
x_test['steps'] = x_test['steps'].apply(' '.join)
x_test['ingredients'] = x_test['ingredients'].apply(' '.join)
x_test
```

```python
all_words = []

for i in range (0, len(x_test['steps'])):
    s = ''
    s += x_test.loc[i,'name'] + x_test.loc[i, 'steps'] + x_test.loc[i, 'ingredients']
    all_words.append(s)
    
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import scipy.sparse
# stop_word ='english' means deleting some common words like the, a,.... in list 'english'



vectorizer = CountVectorizer(stop_words='english', max_features = 150)
X = vectorizer.fit(all_words)
vocab_dict = vectorizer.vocabulary_
# vocab = np.array(vectorizer.get_feature_names())

X_final = vectorizer.transform(all_words)
X_final.shape

# np.savez('CountVec.npz', X_final)
scipy.sparse.save_npz('CountVec.npz', X_final)
```
